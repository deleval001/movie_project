```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import re
from typing import List


# import file
movie_list = pd.read_csv('data/movie_list.csv')

# drop nulls
movie_list_real = movie_list.dropna()
```

```python
# VALERIE ANALYSIS

#VALERIE SUMMARY STATS
v_picks = movie_list_real["Chosen By"]== "Valerie"
v_sum=v_picks.value_counts()


v_rating = movie_list_real["Valerie Rating"]
v_avg= v_rating.mean()
v_avg

v_max= v_rating.max()

v_min = v_rating.min()
v_min

stdvv= v_rating.std()
stdvv

print("Valerie's highest rating is", v_max)
print("Valerie's lowest rating is", v_min)
print("Valerie's average rating is", v_avg)
print("Valerie's standard deviation is", stdvv)
```

```python
# Relative Frequency Valerie ratings
movie_list_real["Valerie Rating"].value_counts(normalize=False)
```

```python
# Valerie IQR
seventy_fifth= v_rating.quantile(0.75)
twenty_fifth= v_rating.quantile(0.25)

v_rating_IQR = seventy_fifth - twenty_fifth
print(seventy_fifth,twenty_fifth)
print("The IQR for Valerie's rating is", v_rating_IQR)
```

```python
# VALERIE VISUALIZATIONS
#Boxplot for ALL movies valerie rated 

sns.boxplot(data= movie_list_real, y=v_rating)
plt.show()
```

```python
# Box Plot of valerie pick's ratings
sns.boxplot( x = movie_list_real["Chosen By"]=='Valerie', y=v_rating)
```

```python
# Valerie rating histogram
sns.histplot(data=movie_list_real, x=v_rating)
plt.title('Distribution of Valerie ratings')
plt.xlabel('Scale of 1-10')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```
